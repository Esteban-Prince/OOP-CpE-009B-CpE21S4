```python
from abc import ABC, abstractmethod

class Employee(ABC):
    """Abstract base class for all employees"""

    def __init__(self, emp_id, emp_name, emp_address):
        self.__emp_id = emp_id
        self.__emp_name = emp_name
        self.__emp_address = emp_address

    @property
    def emp_id(self):
        return self.__emp_id

    @property
    def emp_name(self):
        return self.__emp_name

    @property
    def emp_address(self):
        return self.__emp_address

    @abstractmethod
    def calculate_salary(self):
        """Abstract method to calculate salary"""
        pass


class FullTimeEmployee(Employee):
    """Represents a full-time employee"""

    def __init__(self, emp_id, emp_name, emp_address, allowance, rate):
        super().__init__(emp_id, emp_name, emp_address)
        self.__allowance = allowance
        self.__rate = rate

    @property
    def allowance(self):
        return self.__allowance

    @property
    def rate(self):
        return self.__rate

    def calculate_salary(self):
        """Calculates salary for a full-time employee"""
        return (self.__rate * 22) + self.__allowance # Assuming 22 working days


class PartTimeEmployee(Employee):
    """Represents a part-time employee"""

    def __init__(self, emp_id, emp_name, emp_address, rate):
        super().__init__(emp_id, emp_name, emp_address)
        self.__rate = rate

    @property
    def rate(self):
        return self.__rate

    def calculate_salary(self):
        """Calculates salary for a part-time employee"""
        return self.__rate * 15 # Assuming 15 working days


class Salary:
    """Represents an employee's salary record"""

    def __init__(self, salary_id, salary, cut_off_date, days_of_work):
        self.__salary_id = salary_id
        self.__salary = salary
        self.__cut_off_date = cut_off_date
        self.__days_of_work = days_of_work

    @property
    def salary_id(self):
        return self.__salary_id

    @property
    def salary(self):
        return self.__salary

    @property
    def cut_off_date(self):
        return self.__cut_off_date

    @property
    def days_of_work(self):
        return self.__days_of_work


# Example usage
full_time_employee = FullTimeEmployee(1, "Mr. Kupido", "123 Main Street", 500, 100)
part_time_employee = PartTimeEmployee(2, "Ms. Kupido", "456 Oak Avenue", 80)

# Calculate and store salary
full_time_salary = full_time_employee.calculate_salary()
salary_record = Salary(1, full_time_salary, "2024-01-31", 22)

print(f"Full-time Employee Salary: {salary_record.salary}")
print(f"Part-time Employee Salary: {part_time_employee.calculate_salary()}")
print("Full-time Employee Rate:", {full_time_employee.rate})
print("Part-time Employee Rate:", {part_time_employee.rate})


print("\n")
print("----------Employee Information----------")


print(f"Employee ID: {full_time_employee.emp_id}")
print(f"Employee Name: {full_time_employee.emp_name}")
print(f"Employee Address: {full_time_employee.emp_address}")
print(f"Allowance: {full_time_employee.allowance}")



print(f"Salary: {salary_record.salary}")
print(f"Cut-off Date: {salary_record.cut_off_date}")
print(f"Days of Work: {salary_record.days_of_work}")

print("-----------------------------------------")




```
